### 먼저 통계는 두가지로 나뉜다

기술통계와 추론통계

- 기술통계 : descriptive statistic 
    - 샘플에 대한 특성인 평균, 표준편차, 중위수, 최빈값, 그래프, 왜도, 첨도 등을 구하는 것
    - 

- 추측통계 : 
    - 모수추정 : 모집단의 특성인 모수(평균, 분산 등)을 분석하여 모집단을 추론
    - 가설검정 : 대상 집단에 대한 가설을 설정한 후 그 가설이 옳은지 판단
    - 예측    : 미래의 불확실성을 해결해 효율적인 의사결정을 하기 위해 활용 (회귀분석, 시계열 분석등)

- 기술 통계치(Descriptive Statistics)란?
    - 기술통계치는 두가지로 나뉩니다. < 중앙 집중화 와 분산화 >
    - 중앙 집중화는 평균 중앙값 최빈값 등을 말하고
    - 분산화는 퀀타일 (사분위수), 표준편차 등을 의미합니다.
    - 측정이나 실험에서 수집한 자료(data)의 정리, 요약, 해석 ,표현등을 통해 자료의 특성을 규명하는 통계적 방법
    - 즉, 수집한 데이터를 요약, 묘사하는 통계기법
    - 분산 : 데이터가 얼마나 퍼져있는가  ⇒ 편차를 제곱해 평균내버린 것
    
- 추리 통계치(Inferential Statistics)
    - infer : 추론하다 예측하다. 추론하고 예측하는 통계기법
    - 설문조사로 기호 1번이 당선될것이다 같은 기법

검증과 검정의 차이

- 검증[檢證] 은 법률용어에서 "법관이 다툼이 있는 사실에 대하여 판단의 기초자료를 확보하기 위해 자기의 감각을 스스로 실험하는 증거조사"를 의미한다. 즉 어떤 사실의 진위를 판단하기 위해 증거자료를 수집하여 옳고 그름을 증명하는 것이다.

- 검정은 [檢正]은 잘 조사하여 바르게 하는 일을 뜻한다. 영어로 test 
연구자가 현상을 관찰하고 어떤 변수들간의 인과관계 또는 상관관계를 설정한 문장을 가설이라고 한다. 사회과학의 연구방법론으로 가장 많이 쓰이는 것이 가설을 설정하고 데이터를 수집하여 그 가설이 맞는지 검정(test)해보는 것이다. 
경영서적이나 논문에서 가장 많이 틀리는 것이 "가설검증"이라는 말이다. 이는 "가설검정"이 맞다. 한국통계학회에서 발간한 통계용어사전에도 Hypothesis Test는 "가설검정"이라고 나와있다.

#### P- value

- 갑자기 뜬금 p-value가 등장했다.
- 일단 외우고 가자 P값이 0.05보다 작으면 귀무가설 기각! 
- 이 p값은 저기 빨간 부분의 넓이보다 작은지 큰지 그러한 우연성의 넓이가 얼마나 되는지를 알려준다. 
- 즉 p값도 넓이라고 보면된다 => 왜냐면 이것도 확률이니까! 


그림은 P-value 요약짤

- p-value가 0.05(유의수준)보다 작다 -> 우리가 인정하기로 한 귀무가설보다 더 극단적이다.
- '사람들은 길을 가다가 번개를 맞고, 갑자기 드래곤이 나타나서 잡아먹혀 죽는다' 라는 가설을 검증한다고 했을 때 **0.000001%**의 확률로 해당 사건이 실제로 벌어질 수 있지만 (= P-value) 그렇다고 해서 모든 사람들이 드래곤을 대비한 비싼 미스릴 갑옷을 살 필요는 없다.(= 비용이 많이 들기 때문에)
- 그런 사건이 진짜 일어난다고 가정해 버린다면 너무 지출해야할 비용이 커진다. '그런 사건은 일어나지 않는다'라는 가설을 채택하고 각자 잘 사는게 맞는 결론이 나온다.
