```python
a = sns.load_dataset('taxis')
a
```

- 'fare', 'tip', 'distance' 연속형변수
- 'passengers' = 카테고리형 변수 

```python
from statsmodels.multivariate.manova import MANOVA

host = a[['passengers', 'fare', 'tip', 'distance']]
print(MANOVA.from_formula('fare + tip + distance~passengers', data = host).mv_test())
```

- 위가 상수항
- 아래가 독립변수
- 위는 통계적으로 유의미함
